```python
for root, dirs, files in os.walk(path_uni):
    for file in files:
        if file.endswith(".label"):
            checker = os.path.join(root, file)
            labels = np.fromfile(checker, dtype='uint32')

            newArray = copy(labels)

            (unique, counts) = np.unique(newArray, return_counts=True)
            frequencies = np.asarray((unique, counts)).T
            print(frequencies)

            for k, v in val_dict.items(): newArray[labels==k] = v

            newArray[newArray >= 20] = 0

            (uniq, count) = np.unique(newArray, return_counts=True)
            frequenci = np.asarray((uniq, count)).T
            print(frequenci)

            newArray.astype('uint32').tofile(checker)

            del labels
            del newArray
            del unique
            del counts
            del frequencies
            
```
